```python
df11[['year','Month']] = df11['month'].str.split('-',expand = True)
df11['block'] = df11['block'].str.split(r'(\D)', expand = True)[0].astype('int64')
df11[['storey_from','storey_to']] = df11['storey_range'].str.split('TO',expand = True)
df11['flat_type'] = df11['flat_type'].replace({'EXECUTIVE': 6, 'MULTI-GENERATION': 10})
df11['flat_type'] = df11['flat_type'].astype(str).str.extract('(\d+)')
```

```python
En = OrdinalEncoder()
df11['street_name'] = En.fit_transform(df11[['street_name']])
df11['flat_model'] = En.fit_transform(df11[['flat_model']])
df11['town'] = En.fit_transform(df11[['town']])
```

```python
df11.drop('storey_range',axis = 1, inplace = True)
df11.drop('month',axis = 1, inplace = True)
```

```python

df11['Month'] = df11['Month'].astype('int64')
df11['town'] = df11['town'].astype('int64')
df11['floor_area_sqm'] = df11['floor_area_sqm'].astype('int64')
df11['flat_model'] = df11['flat_model'].astype('int64')
df11['resale_price'] = df11['resale_price'].astype('int64')
df11['year'] = df11['year'].astype('int64')
df11['storey_from'] = df11['storey_from'].astype('int64')
df11['storey_to'] = df11['storey_to'].astype('int64')
df11['flat_type'] = df11['flat_type'].astype('int64')
```

```python
x = df11.drop('resale_price' , axis = 1)
y = df11['resale_price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.15)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x.head(1)
```

```python
y.head(1)
```

```python
result = DecisionTreeRegressor()
model = result.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print("Mean Absolute Error = ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error =", metrics.mean_squared_error(y_test,y_pred))
print("R2 score=", metrics.r2_score(y_test,y_pred))
```

```python
result1 = RandomForestRegressor()
model1 = result1.fit(x_train,y_train)
y_pred1 = model1.predict(x_test)
```

```python
print("Mean Absolute Error", metrics.mean_absolute_error(y_test,y_pred1))
print("Mean Squared Error", metrics.mean_squared_error(y_test,y_pred1))
print("R2 score =", metrics.r2_score(y_test,y_pred1))
```
